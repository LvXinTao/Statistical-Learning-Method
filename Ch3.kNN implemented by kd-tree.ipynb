```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#这里实现2维数据

## 节点类
class Node(object):
    def __init__(self,data,sp=0,left=None,right=None):
        self.data=data
        self.sp=sp
        self.left=left
        self.right=right
    
    def __lt__(self,other):
        return self.data<other.data
        
        
```

```python
class kdTree(object):
    def __init__(self,data):
        self.dim=data.shape[1]
        self.root=self.createTree(data,0)
        self.nearesr_node=None
        self.nearest_dist=np.inf
        
    def createTree(self,dataset,sp):
        '''
        根据训练集创建kdTree
        '''
        if len(dataset)==0:
            return None
        
        ## 按特征列排序
        dataset_sorted=dataset[np.argsort(dataset[:,sp])] 
        ## 获取中位数索引
        mid=len(dataset)//2
        ## 生成节点
        left=self.createTree(dataset_sorted[:mid],(sp+1)%self.dim)
        right=self.createTree(dataset_sorted[mid+1:],(sp+1)%self.dim)
        parentNode=Node(dataset_sorted[mid],sp,left,right)
        
        return parentNode
    
    def nearest(self,x):
        def visit(node):
            if node is not None:
                dis=node.data[node.sp]-x[node.sp]
                ## 访问子节点
                visit(node.left if dis>0 else node.right)
                ## 查看当前节点到目标节点的二范数距离
                curr_dis=np.linalg.norm(x-node.data,2)
                ## 更新节点
                if curr_dis<self.nearest_dist:
                    self.nearest_dist=curr_dis
                    self.nearesr_node=node
                ## 如果目标节点到当前节点距离超过当前超平面，则需要到另一边查找
                if self.nearest_dist>abs(dis):
                    visit(node.left if dis<0 else node.right)
                    
        node=self.root
        visit(node)
        return self.nearesr_node,self.nearest_dist
```

```python
## test
data=np.array([[2,3],[5,4],[9,6],[4,7],[8,1],[7,2]])
kdtree=kdTree(data)
node,dist=kdtree.nearest(np.array([6,5]))
print(node.data[0],node.data[1],dist)
```
