```python
import numpy as np
```

```python
class Perceptron(object):
    def __init__(self):
        self.w=np.zeros(5)
        self.b=0
        self.dim=5
        self.lr=0.01
        self.epoches=50
        
    def train(self,data):
        datalen=len(data)
        y=data[:,-1]
        x=data[:,:-1]
        for i in range(self.epoches):
            ## 随机梯度下降更新参数，每次选取一个误分类点，使用其数据进行参数更新
            sgd=np.random.randint(datalen)
            y_s=y[sgd]
            x_s=x[sgd]
            res=(np.dot(x_s,self.w)+self.b)*y_s
            if res<=0:
                self.w=self.w+self.lr*y_s*x_s
                self.b=self.b+self.lr*y_s
            else:
                continue
    
    def discriminate(self,data):
        return np.dot(self.w,data)+self.b
            
```

```python
data=np.array([[1,2,3,4,5,1],
              [2,2,2,2,2,-1],
              [2,3,4,5,6,1],
              [3,3,3,3,3,-1],
              [3,4,5,6,7,1],
              [4,4,4,4,4,-1]
              ])
```

```python
p=Perceptron()
p.train(data)
```

```python
p.discriminate([5,5,5,5,5])
```
